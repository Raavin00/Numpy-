```python
# Most of the Matplotlib utilities lies under the pypot submodule.
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,6])
y=np.array([0,250])
plt.plot(x,y)
plt.show()
```

```python
#plotting without line
x1=np.array([11,22])
y1=np.array([11,22])
plt.plot(x1,y1,'o')
plt.show()
```

```python
#Multiple Points
x2=np.array([1,2,6,8])
y2=np.array([8,4,1,10])
plt.plot(x2,y2)
plt.show()

#Default X-points
#If we don't mention x-axis values ,it takes default values([0,1,2,3,4,.......]) based on length of y-axis.
y2=np.array([8,4,1,10])
plt.plot(y2)
plt.show()
#Defalut values are taken as ([0.0,0.5,1.0,1.5,2,0,.......])
```

```python
# Matplotlib Markers
# We can use MARKER (keyword argument ) to specify a marker
x3=np.array([11,22,33,44,55])
y3=np.array([0,122,44,55,11])
plt.plot(x3,y3,marker='*')
plt.plot()
# We can use diffenent markers like o,*,x,X,+,P(plus(filled)),s(square),D(diamond),d(diamond,thin),p(pentagon),
#H,h(hexagon),(V,<,>,^,triangle up down left right)

#Format Strings(solid line'-',Dotted line ':',Dashed line '--',Dahed/dotted_line '-.')
y4=np.array([12,22,3,11,21])
plt.plot(y4,':')
```

```python
# Marker Size(ms)
y6=np.array([7,8,2,1])
plt.plot(y6,marker='v', ms=10)
plt.show()
```

```python
#Marker_edge_color(shortly reffered as mec)
y0=np.array([19,98,70,68])
plt.plot(y0,marker='H',ms=10,mec='r')
plt.show()
```

```python
#Maker_face_color(mfc)
y11=np.array([12,5,4,66])
plt.plot(y11,marker='o',ms=15,mfc='k')
plt.show()
```

```python
#LineStyle(shortly as ls) eg : dotted ,dashed,solid,dashdot,
#LineColor 
#LineWidth

y22=np.array([3,8,1,10])
plt.plot(y22,linestyle='solid' ,color='hotpink' ,linewidth='100')
plt.show()
```

```python
#Multiple Lines
y33=np.array([38,1,10,23,21])
y44=np.array([11,65,43,21])
plt.plot(y33)
plt.plot(y44)
plt.show()
```

```python
# Matplotplotlib Labels and Titles
# with pyplot we can use the xlabel() ylabel() functions to label the x-axis and y-axis
# We can also add Title to the Graph

xx=np.array([15,33,64,98])
yy=np.array([21,43,65,87])
plt.plot(xx,yy)
plt.xlabel("Avengers Team")
plt.ylabel("Scavengers Team")
#plt.title()
plt.title("War Strength")
plt.show()
```

```python
# fontdict parameter to set font properties for the labels and titles
yyy=np.array([22,33,65,87,333])
plt.plot(yyy)
font1={'family':'serif','color':'black','size':10}
font2={'family':'serif','color':'blue','size':10}
plt.title("Gender",fontdict = font1)
plt.xlabel("Boys",fontdict=font1)
plt.ylabel("Girls",fontdict=font2)
plt.show()
```

```python
#Position the Title using loc() parameter
yyy=np.array([22,33,65,87,333])
plt.plot(yyy)
font1={'family':'serif','color':'black','size':10}
font2={'family':'serif','color':'blue','size':10}
plt.title("Gender",fontdict = font1,loc='right')
plt.xlabel("Boys",fontdict=font1)
plt.ylabel("Girls",fontdict=font2)
plt.show()
```

```python
# Adding Grid Lines using grid() *function
yyy=np.array([22,33,65,87,333])
plt.plot(yyy)
font1={'family':'serif','color':'hotpink','size':10}
font2={'family':'serif','color':'hotpink','size':10}
plt.title("Gender",fontdict = font1)
plt.xlabel("Boys",fontdict=font1)
plt.ylabel("Girls",fontdict=font2)
plt.grid()
plt.show()
```

```python
# Size 
a=np.array([1,3,7,8,10,3,6,8,2,9,22,16])
b=np.array([4,5,9,10,11,12,9,8,12,17,18,7])
c=np.array([0,1222,123,441,65,623,111,135,336,327,23,29])

plt.scatter(a,b,s=c)
plt.show()

# Alpha
# you can adjust the transparency of the dots with alpha argument

a=np.array([1,3,7,8,10,3,6,8,2,9,22,16])
b=np.array([4,5,9,10,11,12,9,8,12,17,18,7])
c=np.array([0,1222,123,441,65,623,111,135,336,327,23,29])

plt.scatter(a,b,s=c,alpha=0.5)
plt.show()
```

```python
# combine color size and Alpha
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
size=100*np.random.randint(100,size=(100))

plt.scatter(x,y,c=colors,s=size,alpha=0.5,cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

```python
# Creating Bars
# with Pyplot you can use the bar() function to draw bar graphs

x=np.array(["A","B","c","D","E"])
y=np.array([5,3,9,1,15])
plt.bar(x,y)
plt.show()
```

```python
# Horizontal Bar Graphs
x=np.array(["R","A","A","V","I","N"])
y=np.array([2,5,7,1,9,4])
plt.barh(x,y)
plt.show()

# Bar color
x=np.array(["R","A","A","V","I","N"])
y=np.array([2,5,7,1,9,4])
plt.bar(x,y,color='red')
plt.show()

# Bar Width
x=np.array(["R","A","A","V","I","N"])
y=np.array([2,5,7,1,9,4])
plt.bar(x,y,color='green',width=0.5)
plt.show()

# Bar Height
x=np.array(["R","A","A","V","I","N"])
y=np.array([2,5,7,1,9,4])
plt.barh(x,y,height=0.1)
plt.show()
```

```python
# Matplotlib Pie Charts
y=np.array([25,34,12,54])
plt.pie(y)
plt.show()

# labels
# you can use labels parameter to labels the each wedge with differnt colors

y=np.array([25,34,12,54,22])
mylabels=["Pineapple","Grapes","Strawberries","Oranges","Dates"]
plt.pie(y,labels=mylabels)
plt.show()

# StartAngle(default angle is 0)
y=np.array([25,34,12,54,22])
mylabels=["Pineapple","Grapes","Strawberries","Oranges","Dates"]
plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

```python
# Explode
# we can exolode the wedge by using the parameter EXplode()

y=np.array([25,34,12,54,22])
mylabels=["Pineapple","Grapes","Strawberries","Oranges","Dates"]
myexplode=[0.2,0.3,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

```python
# Shadow

y=np.array([25,34,12,54,22])
mylabels=["Pineapple","Grapes","Strawberries","Oranges","Dates"]
myexplode=[0.2,0.3,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

```python
# You can set the colors of each widge with the colors parameter

y=np.array([25,34,12,54,22])
mylabels=["Pineapple","Grapes","Strawberries","Oranges","Dates"]
mycolors=["black","pink","green","violet","brown"]
plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

```python
#Legend

y=np.array([25,34,12,54,22])
mylabels=["Pineapple","Grapes","Strawberries","Oranges","Dates"]
plt.pie(y,labels=mylabels)
plt.legend()
plt.show()

# Legend with Header
y=np.array([25,34,12,54,22])
mylabels=["Pineapple","Grapes","Strawberries","Oranges","Dates"]
plt.pie(y,labels=mylabels)
plt.legend(title="Five Fruits :")
plt.show()
```
